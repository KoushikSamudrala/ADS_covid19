# Data Preparation

# John Hopkins Github csv data

```python
data_path='../data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
pd_raw=pd.read_csv(data_path)
```

```python
#analysing the data
pd_raw.head()
```

```python
#we want to have date columns as a separate vector index and thus extracting columns only with dates
time_idx=pd_raw.columns[4:]
```

```python
#we can make this index as column to new dataframe
df_plot=pd.DataFrame({'date':time_idx})
df_plot.head()
```

```python
#inspect individual countries in the dataframe
pd_raw['Country/Region'].unique()
```

```python
#to inspect individual countries daily infections
pd_raw[pd_raw['Country/Region']=='Germany'].iloc[:,4::].sum(axis=0)[:4]
```

```python
#we will create a smaller list of countries among all countries
country_list=['Italy','US','Spain','India','Germany']
```

```python
for each in country_list:
    df_plot[each]=np.array(pd_raw[pd_raw['Country/Region']==each].iloc[:,4::].sum(axis=0)) #np.array to create values into dataframe instead of NaN
```

```python
df_plot
```

```python
#our first result
df_plot.set_index('date').plot()
```

# Data type Date

```python
#lets look at data once
df_plot.head()
```

Here date format is not as per our requirement standard (ISO)

```python
#to convert string to a date object
datetime.strptime(df_plot.date[0],"%m/%d/%y")
```

```python
#convert every date string to datetime object
time_idx=[datetime.strptime(each,"%m/%d/%y") for each in df_plot.date]
```

```python
time_idx[0:5]
```

```python
#convert back to string format ISO norm
time_str=[each.strftime('%Y-%m-%d') for each in time_idx]
time_str[0:5]
```

```python
#push time-idx into date column and look at data type
df_plot['date']=time_idx
type(df_plot['date'][0])
```

```python
df_plot.head()
```

```python
#save our model to processed
df_plot.to_csv('../data/processed/COVID_small_flat_table.csv',sep=';',index=False)
```

# Relational Data Model

```python
#for big models
#we will start again with our original dataset
data_path='../data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
pd_raw=pd.read_csv(data_path)
pd_raw.head()
```

if we observe data, lat and longitude doesnt require here as they are of no interest to us and these values wont change over time. whereas values for dates do change over time. so we have to split these two types.

```python
#renaming the columns
pd_data_base=pd_raw.rename(columns={'Country/Region':'country','Province/State':'state'})
```

```python
#dropping of unnecessary columns
pd_data_base=pd_data_base.drop(['Lat','Long'],axis=1)
pd_data_base.head()
```

```python
#we can push state and country as index considering primary keys
pd_data_base.set_index(['state','country'])
```

```python
#we can switch rows and column index
test_pd=pd_data_base.set_index(['state','country']).T
```

```python
test_pd.head()
```

```python
#we can change this multilevel columns to row perspective i.e change state or country column as row by stack command
test_pd.stack(level=0) #level=1 for country as row
```

```python
#to set everything as row level and renaming columns
test_pd.stack(level=[0,1]).reset_index().rename(columns={'level 0':'date',0:'confirmed cases'})
```
